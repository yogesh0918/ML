## Train Test Splitting

```python
import numpy as np
def split_train_test(data,test_ratio):
    
    np.random.seed(42) ## This will fixed randomizing data again and again. Only 1st time data will randomlly selected.
    
    shuffled = np.random.permutation(len(data)) ## This will randomally selects data from dataset.
    
    test_size = int(len(data) * test_ratio)
    train_indices = shuffled[test_size:]
    test_indices = shuffled[:test_size]
    return data.iloc[train_indices], data.iloc[test_indices]
```

```python
train_data, test_data = split_train_test(housing,0.2)
```

```python
print(f'''Number of rows in train dataset: {len(train_data)}
Number of rows in test dataset: {len(test_data)}''')
```

# Feature Scalling

Feature Scalling has Primirily 2 types,
1. Normalization - (MinMax Scalling)
    Normalization = (value - min) / (max - min)
    
    Sklearn has class as MinMaxScaler 
    
2. Standardization
    Standardization = (value - mean) / (std)
    
    Sklearn has class as StandardScaler   
    
Using Standardization will be helpfull, beacause it uses mean and standard deviation; Even if max or min value eleminate from data will not affect the data that much; But if we use Normalization and removed min or max value then Data will <b>Drasstically</b>
affects.

# Scikit-learn Design

## Create Pipeline

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
```

```python
my_pipeline = Pipeline([
    ("imputer",SimpleImputer(strategy="median")),
    ("std_Scaler",StandardScaler()),
])
```

```python
Housing_Data = my_pipeline.fit_transform(housing)
Housing_Data
```

```python
Housing_Data.shape
```

## Selecting a desired model for Dragon Real Estate 

```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor

# model = LinearRegression()
# model = DecisionTreeRegressor()
model = RandomForestRegressor()
model.fit(Housing_Data,housing_labels)
```

```python
some_data = housing.iloc[:5]
some_labels = housing_labels.iloc[:5]
prep_data = my_pipeline.transform(some_data)
model.predict(prep_data)
```

# Evaluating model

```python
import numpy as np
np.array(some_labels)
```

```python
from sklearn.metrics import mean_squared_error
housing_predict = model.predict(Housing_Data)
mean_sq_error = mean_squared_error(housing_labels,housing_predict)
root_mean_sqrt_error = np.sqrt(mean_sq_error)
root_mean_sqrt_error
```

# Using better evaluation technique - Cross validation

```python
from sklearn.model_selection import cross_val_score
score = cross_val_score(model,Housing_Data,housing_labels,scoring="neg_mean_squared_error",cv=10)
rmse_scores = np.sqrt(- score)
```

```python
rmse_scores
```

```python
def print_value(scores,model)-> str: 
    return(f"Model:{str(model)[:-2]}\nMean: {scores.mean()}\nStandard Deviation: {scores.std()}")
```

```python
Model_Output = print_value(rmse_scores,model)
Model_Output
# print(Model_Output)
with open('Models_Output.txt', 'a') as the_file:
    the_file.write("\n\n"+Model_Output)
```

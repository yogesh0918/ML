## Train Test Splitting

```python
import numpy as np
def split_train_test(data,test_ratio):
    
    np.random.seed(42) ## This will fixed randomizing data again and again. Only 1st time data will randomlly selected.
    
    shuffled = np.random.permutation(len(data)) ## This will randomally selects data from dataset.
    
    test_size = int(len(data) * test_ratio)
    train_indices = shuffled[test_size:]
    test_indices = shuffled[:test_size]
    return data.iloc[train_indices], data.iloc[test_indices]
```

```python
train_data, test_data = split_train_test(housing,0.2)
```

```python
print(f'''Number of rows in train dataset: {len(train_data)}
Number of rows in train dataset: {len(test_data)}''')
```

```python
from sklearn.model_selection import train_test_split
train, test = train_test_split(housing, test_size = .2, random_state = 42)
```

```python
print(f'''Number of rows in train dataset: {len(train)}
Number of rows in test dataset: {len(test)}''')
```

```python
from sklearn.model_selection import StratifiedShuffleSplit
split = StratifiedShuffleSplit(n_splits = 1,test_size = .2,random_state= 42)

for test_index, train_index in split.split(housing,housing["CHAS"]):
    strat_test_index = housing.loc[test_index]
    strat_train_index = housing.loc[train_index]
```
